```python
# Filling the Null values
```

```python
import pandas as pd
df=pd.read_csv('/content/testingdata(Sheet1).csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#sometimes we can't remove data.because sometimes ur physics mark is not there, if we are removing data our chem, maths mark also going
df.fillna(0) # it will fill all the null values with 0
```

```python
df.fillna(100)
```

```python
df.fillna({'no':'none','values':'0','tags':'-'})
#  here  null values from diff columns filling with different fillings
```

```python
# for filling null values 2 methods are there - 1. forward fill(ffill) 2.backwardfill(bfill)
#forward fill-filling null values with previous value, here we will define methpd with the name as ffill
df.fillna(method='ffill')   #eg:previous one(top one) is 44 ,then this null also written as 44
#default value is row , axis=0, so it will take previous row value
```

```python
df.fillna(method='ffill',axis=1)
#null value filled with previous values in columns
```

```python
df['no'].fillna(value=df['no'].mean()) # here the null values in 'no' filling wih the mean value of no.
```

```python
df.fillna(method='bfill') # null value printing with coming value
```

```python
df.fillna(method='bfill',inplace=True)
df # here original database will lost, it will replace with current database.
```
