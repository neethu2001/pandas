```python
import pandas as pd
df=pd.read_csv('/content/testingdata(Sheet1).csv') #index_col=['roll no'] - here we can add this and chnage roll no to index
df
```

```python
df.loc[1] #location according to index
```

```python
df.loc[5]
```

```python
df.loc[[5,6,7,8]] # loc array of indexes
```

```python
df.loc[5,'no']
```

```python
df.loc[5:10,'values']
```

```python
df.loc[df['no']<50]
```

```python
df.loc[df['no']>40] # more than 40
```

```python
df.loc[df['no']>40,['values']]
# here the person who have more than 40 no, need to show only the values
```

#iloc- index location

```python
# in loc heading is taking as 0th index.
# but in iloc 0th index is starting from values
```

```python
df.iloc[0]
```

```python
df.iloc[[0]] # here we will get the array of 0th index
```

```python
df.iloc[[0,1,2,3]]
```

```python
df.iloc[:,0] #show all the rows from the oth column
```

```python
df.iloc[:,2] # print all rows in column 2
```

```python
df.iloc[0:5,1] # from 0 to 5 rows in 1st index column
```

```python
df.iloc[0:5,1:5] # print rows from 0 to 5, columns from 1 to 5
```
