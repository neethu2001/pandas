```python
import pandas as pd
```

```python
df1=pd.DataFrame({'Roll no':[1,2,3,4,5],'maths':[46,78,42,68,41],'physics':[42,39,68,63,72]})
df2=pd.DataFrame({'Roll no':[6,7,8,9,10],'maths':[56,79,40,48,71],'physics':[56,49,54,78,41]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
#ignore_index=True means it will continue the index.. it will continue the index like 5,6.. after the df1 index
```

```python
pd.concat([df1,df2],ignore_index=True, sort=True) #sort the column name
```

```python
df1=pd.DataFrame({'Roll no':[1,2,3,4,5],'chemistry':[46,78,42,68,41],'physics':[42,39,68,63,72]}) # columns in both dataframe are different
df2=pd.DataFrame({'Roll no':[6,7,8,9,10],'maths':[56,79,40,48,71],'physics':[56,49,54,78,41]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],ignore_index=True) #till index 4 maths marks will be null, after that chemistry mark will be null
```
