*pandas dataframe -Data sets in Pandas are usually multi-dimensional tables, called DataFrames
Series is like a column, a DataFrame is the whole table.*

```python
import pandas as pd
```

```python
df=pd.DataFrame()
print(df)
```

```python
# DATAFRAME USING LIST

lst=[1,2,3,4,5,6]
df=pd.DataFrame(lst)
print(df)
```

```python

lst=[[1,2,3,4,5,6],[11,12,13,14,15]]
df=pd.DataFrame(lst)
print(df)
```

```python
a=[{'a':5,'b':6,'c':7,'d':8},
 {'a':3,'b':1,'c':2,'d':9}] # dictionary keys rep columns names
df=pd.DataFrame(a) # a,b,c,d are the keys
df
```

```python
b={'roll no':pd.Series([1,2,3]),
   'maths':pd.Series([45,43,23]),
   'physics':pd.Series([38,40,48])}
   # instead of giving the values we are using series
df=pd.DataFrame(b)
df
```
