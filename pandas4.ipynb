```python
#DROPPING COLUMN WITH NULL VALUES(nan)
```

```python
df3=df.dropna(axis=1) # here we are dealing with columns
df3
```

```python
# here we are dealing with rows
df4=df.dropna(axis=0)
df
```

```python
df5=df.dropna(how='any') # any value is null, then drop that row
df5
```

```python
df6=df.dropna(how='all') # if all the row values is null
df6
```

```python
df.dropna(inplace=True)
```
