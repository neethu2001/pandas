#Merge() Function

```python
df1=pd.DataFrame({'flower':[1,2,3,4,5],'no':[34,67,89,35,23]})
df1
```

```python
df2=pd.DataFrame({'flower':[1,2,3,4,5],'tags':[67,4,11,35,23]})
df2
```

```python
pd.merge(df1,df2,on='flower') # merge tthe two dataframes using a common values as flower
```

```python
pd.merge(df2,df1,on='flower') #df2 is printing first, then df1
```

```python
df3=pd.DataFrame({'flower':[1,2,3,6,7],'tags':[67,4,11,68,23]})
df4=pd.DataFrame({'flower':[1,2,3,8,9],'no':[23,45,121,60,73]})
pd.merge(df3,df4, on='flower') # it will merge only on the common flowe index as 1,2,3
```

```python
pd.merge(df3,df4,how='right') # it is taking the values based df3- it will take only 1,2,3,6,7
#in 6,7  value of 'no' is none. because that data is not provided
```

```python
pd.merge(df3,df4,how='left')
```

```python
pd.merge(df3,df4,how='outer')# will take all the values
```
