```python
#pivot table
import pandas as pd
df=pd.read_csv('/content/testingdata(Sheet1) (2).csv')
df
```

```python
pd.pivot_table(df,index='Branch') # creating pivot table based on branch
#branch is taking as index , also it is calculating the average of each and every column
```

```python
pd.pivot_table(df,index='Branch',aggfunc='sum') #by default it is taking mean.
# so i changed into sum
```

```python
pd.pivot_table(df,index='Branch',aggfunc='max') # here it is taking max value
```

```python
pd.pivot_table(df,index='Branch',columns='tags')
```
